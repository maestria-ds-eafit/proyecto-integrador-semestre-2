```python
import pandas as pd
```

```python
df = pd.read_csv("counts_per_category.csv")
df.head()
```

Obtenemos el total de todos los registros


```python
total = df["count"].sum(axis=0)
total
```

Calculamos una nueva columna que sea la contribución de cada categoría al total en términos de proporción


```python
df["proportion"] = df["count"] / total
df
```

```python
proportions = df[["category", "proportion"]].to_dict(orient="records")
proportions
```

```python
result_dict = {}

# Iterate through the list and create the dictionary
for item in proportions:
    category = item["category"]
    proportion = item["proportion"]
    result_dict[category] = proportion

result_dict
```

`n` es el número de muestras que queremos obtener


```python
n = 1e6
```

```python
weights_dict = {}

for key, item in result_dict.items():
    total_in_sample = item * n
    weight = total_in_sample / df[df["category"] == key]["count"].values[0]
    weights_dict[key] = weight

weights_dict
```
